# Automatic Differentiation in PyTorch

## A glance at PyTorch?

- PyTorch is a Deep Learning library/framework in Python
- PyTorch with tensors is similar to Numpy with multi-dimentional arrays, but has more convenience functions for deep learning
- PyTorch is based on Torch7 (a famous deep learning around 5 or 10 years ago), which was based on Lua (kinda Python but for C files) and inspired by Lush
- PyTorch started in 2016 (gradually replaced and written again for being compatible with Python)
- Focuses on flexibiliy and minimizing cognitive overhead
- PyTorch is written in C++ and CUDA (CUDA is like C++ for GPU)
- Dynamic nature of autograd API inspired by [Chainer](https://docs.chainer.org/en/stable/glance.html)


## Core features:
1. Automatic differentiation
2. Dynamic compuation graphs
3. Numpy integration

# Autograd Example

```python
%load_ext watermark
%watermark -a "Duc Nguyen Huu" -v -p torch
```

```python
import torch
from torch.autograd import grad
import torch.nn.functional as F
from IPython.display import Image
```

```python
# create tensors
x = torch.tensor([3.])
w = torch.tensor([2.], requires_grad=True)
b = torch.tensor([1.], requires_grad=True)
```

```python
# activation function is ReLU
a = F.relu(x*w + b) 
a
```

```python
Image("relu's_computation_graph.PNG")
Image('results.PNG')
Image('ReLU.PNG')
```

```python
# compute the gradient of a with respect to w
grad(a, w, retain_graph=True)
```

```python
# compute the gradient of a with respect to w
grad(a, b)
```

# Implement ReLU function

```python
def my_relu(z):
    if z > 0:
        return z
    else:
        z[:] = 0 
        return z
    
a = my_relu(x*w + b)
grad(a, w)
```

- Using PyTorch functions are much more efficient for real-world problems
